```python
# Removing the outlier

def remove_outlier(column_name,df):
    sorted(column_name)
    Q1,Q3 = np.percentile(column_name, [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    
    outliers = df[(column_name < lower_range) | (column_name > upper_range)]
    
    print(outliers)
    return df.drop(outliers.index, inplace=True)
```

```python
def Remove_Outlier_Indices(df):
    """ returns a df of boolean values to used to index a trimmed table in the 1.5*IQR range"""
    if isinstance(df,pd.core.frame.DataFrame):
        Q1 = df.quantile(0.15)
        Q3 = df.quantile(0.85)
        IQR = Q3 - Q1
        trueList = ~((df < (Q1 - 1.5*IQR)) |(df > (Q3 + 1.5*IQR)))
    else:
        raise TypeError
    return trueList
```
