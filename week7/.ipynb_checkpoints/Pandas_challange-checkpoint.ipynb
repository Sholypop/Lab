```python
def split (df):
    df['country'] = df['country'].astype('category')
    df['country_encode'] = df['country'].cat.codes
    ma = df['country_encode'].max()
    cal = ma/2
    df1 = df[df['country_encode'] > cal ]
    df2 = df[df['country_encode'] <= cal ]
    print(df1)
    print()
    print(df2)
```
