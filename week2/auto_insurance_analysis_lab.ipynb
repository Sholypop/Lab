```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv('file1.csv')
df_1.head()
```

```python
df_2 = pd.read_csv('file2.csv')
df_2.head()
```

```python
df_3 = pd.read_csv('file3.csv')
df_3.head()
```

```python
# changing all columns to lower case
def lower_case(data):
    data.columns=[i.lower() for i in data.columns]
    return data

lower_case(df_1)
lower_case(df_2)
lower_case(df_3)
```

```python
# changing st to state
df_1 = df_1.rename(columns={'st':'state'})
```

```python
df_2 = df_2.rename(columns={'st':'state'})
```

```python
df_1.columns
```

```python
df_2.columns
```

```python
df_3.columns
```

```python
df_1 = df_1[['customer', 'state', 'gender', 'income', 'customer lifetime value', 
                       'monthly premium auto', 'number of open complaints',
                       'policy type', 'vehicle class', 'total claim amount', 'education']]
df_1.head()
```

```python
df_2 = df_2[['customer', 'state', 'gender', 'income', 'customer lifetime value', 
                       'monthly premium auto', 'number of open complaints',
                       'policy type', 'vehicle class', 'total claim amount', 'education']]
```

```python
df_3 = df_3[['customer', 'state', 'gender', 'income', 'customer lifetime value', 
                       'monthly premium auto', 'number of open complaints',
                       'policy type', 'vehicle class', 'total claim amount', 'education']]
```

```python
full_df = pd.concat([df_1,df_2,df_3], axis = 0)
full_df.head()
```

```python
full_df.shape
```

```python
# romoving duplicate

full_df = full_df.drop_duplicates()
```

```python
full_df.shape
```

```python
# deleting and rearranging columns
full_df.drop(columns=['customer'], inplace=True)
```

```python
full_df.columns
```

```python
# working with data types

full_df
```

```python
#full_df['customer lifetime value'].apply(lambda x: x.str.strip('%') if x.dtype == 'object' else x)
```

```python
full_df['customer lifetime value'] = full_df['customer lifetime value'].astype(str).str.strip('%')
```

```python
full_df
```

```python
full_df.info()
```

```python
#convertinf type to integer
```

```python
full_df['customer lifetime value'] =  pd.to_numeric(full_df['customer lifetime value'], errors='coerce')
full_df['customer lifetime value'] = full_df['customer lifetime value'].fillna(0)
full_df['customer lifetime value'] = full_df['customer lifetime value'].astype('int')
```

```python
full_df.info()
```

```python
full_df
```

```python
full_df['number of open complaints'].unique()
```

```python
full_df['number of open complaints'] = full_df['number of open complaints'].astype(str).str.lstrip('1/')
```

```python
full_df
```

```python
full_df['number of open complaints'] = full_df['number of open complaints'].astype(str).str.replace('/00','')
```

```python
full_df
```

```python
full_df.info()
```

```python
full_df['number of open complaints'] =  pd.to_numeric(full_df['number of open complaints'], errors='coerce')
full_df['number of open complaints'] = full_df['number of open complaints'].fillna(0)
full_df['number of open complaints'] = full_df['number of open complaints'].astype('int')
```
