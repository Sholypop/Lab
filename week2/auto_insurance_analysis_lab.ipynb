```python
# REMOVING OUTLIER :
# Interquartal range will be used to identify and remove outliers. 
```

```python
full_df.size
```

```python
def remove_outlier(column_name,df):
    sorted(column_name)
    Q1,Q3 = np.percentile(column_name, [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    
    outliers = df[(column_name < lower_range) | (column_name > upper_range)]
    
    print(outliers)
    return df.drop(outliers.index, inplace=True)
```

```python
# checking the multicollinearity between the input features. multicollinearity does not include the target feature
```

```python
multicolin = numericals_df[['customer_lifetime_value','income','monthly_premium_auto','months_since_last_claim', 
                            'months_since_policy_inception','number_of_open_complaints','number_of_policies']]
```

```python
## variance_inflation_factor() is used to check for multicolinearity between features

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = pd.DataFrame()
vif_data['features'] = multicolin.columns

# calculating VIF for each feature
vif_data['VIF'] = [variance_inflation_factor(multicolin.values, i) for i in range(len(multicolin.columns))]
print(vif_data)
```

1 - features are not correlated
1 < VIF < 5 - features are moderately correlated
VIF > 5 - features are highly correlated
VIF > 10 - high correlation between features and is cause for concern.

```python
# Heatmap ( checking the correlation)

fig=plt.figure(figsize=(8,5))
sns.heatmap(multicolin.corr(),annot=True)
```
